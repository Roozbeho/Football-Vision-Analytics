```python
from pathlib import Path
import os
from ultralytics import YOLO 
# model_path = Path("models/yolo26m.pt").resolve()
model = YOLO("yolo26m.pt")
DEVICE = "cuda"
yaml_file = "/home/roozbeh/programming/DL/DL_uni/football/football_analytics/data/football-players-detection.v1i.yolov8/data.yaml"

metrics = model.train(
    data=str(yaml_file),
    epochs=100,
    imgsz=640,
    batch=4,
    device=DEVICE,
    
    mosaic=1.0,
    mixup=0.15,
    copy_paste=0.2,

    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,

    translate=0.1,
    scale=0.5,
    fliplr=0.5,

    degrees=5,
    perspective=0.0005,
    
    workers=8, 
    project=str(os.getcwd() + "/runs"),
    name="obj_detector",
    exist_ok=True,
    amp=False,
)

best_weights = Path(model.trainer.best)
print("Metrics:", metrics)
print("Best weights:", best_weights)
```

```python
best_weights = os.getcwd() + "/runs/obj_detector/weights/best.pt"

val_model = YOLO(str(best_weights))
metrics = val_model.val(data=str(yaml_file), device=DEVICE, workers=0)
print(metrics)
```

```python
PREDICT_SOURCE = "/home/roozbeh/programming/DL/DL_uni/football/football_analytics/data/football-players-detection.v1i.yolov8/test/images" 
pred_model = YOLO(str(best_weights))
pred_model.predict(
    source=PREDICT_SOURCE,
    device=DEVICE,
    conf=0.25,
    save=True,
    project=str(os.getcwd() + "/runs"),
    name="predict_custom",
    exist_ok=True,
)
```
